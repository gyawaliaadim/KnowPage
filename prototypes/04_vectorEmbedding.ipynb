```python
# # =========================
# 1. IMPORTS
# =========================
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import json
from sentence_transformers import SentenceTransformer
import numpy as np
import joblib
```

```python


# =========================
# 2. LOAD DATA
# =========================
with open("chunks.json", "r", encoding="utf-8") as f:
    data = json.load(f)

print(f"Loaded chunks: {len(data)}")
```

```python

# =========================
# 3. LOAD MODEL
# =========================
MODEL_NAME = "nomic-ai/nomic-embed-text-v1.5"
model = SentenceTransformer(MODEL_NAME, trust_remote_code=True)
```

```python

# =========================
# 4. EMBEDDING FUNCTION
# =========================
def create_embeddings_batch(texts):
    """
    Efficient batch embedding using SentenceTransformers.
    """
    return model.encode(
        texts,
        batch_size=32,
        show_progress_bar=True,
        normalize_embeddings=True  # important for cosine similarity search
    )
```

```python

# =========================
# 5. PROCESS IN BATCHES
# =========================
BATCH_SIZE = 32

chunks_with_vectors = []

for i in range(0, len(data), BATCH_SIZE):
    batch = data[i:i + BATCH_SIZE]

    texts = [chunk["text"] for chunk in batch]

    embeddings = create_embeddings_batch(texts)

    for chunk, emb in zip(batch, embeddings):
        chunk["embedding"] = emb.tolist()  # numpy → JSON serializable
        chunks_with_vectors.append(chunk)

    print(f"Processed {min(i + BATCH_SIZE, len(data))}/{len(data)}")
```

```python

# =========================
# 6. SAVE OUTPUT
# =========================
with open("chunks_with_vectors.json", "w", encoding="utf-8") as f:
    json.dump(chunks_with_vectors, f, ensure_ascii=False)

print("Embedding pipeline complete 🚀 File saved: chunks_with_vectors.json")
```
